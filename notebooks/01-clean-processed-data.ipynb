This notebook produces a single csv with binary `Abnormal` column for all instances of non-normal status, no lagged attributes and NAs removed.

```python
import pandas as pd
full_data = pd.read_csv("../data/intermediate/forecasts_and_statuses.csv")
full_data.head()
```

```python
selected_cols = [col for col in full_data.columns 
                 if col in ["Date", "Abnormal"] or col.endswith("_1") and col not in ["PWH_1", "PWG_1", "CWWH_1", "CWWG_1", "CWE_1"]]
selected_data = full_data[selected_cols]
selected_data = selected_data[selected_data["Date"] < "2024-01-01"]
selected_data.tail()
```

```python
selected_data.describe()
```

```python
selected_data.isna().sum()
```

```python
del selected_data["AREG_1"]
```

```python
selected_data[selected_data.isna().any(axis=1)]
```

```python
selected_data = selected_data.dropna()
selected_data.to_csv("../data/processed/clean.csv", index=False)
```
