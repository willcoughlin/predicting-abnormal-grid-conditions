### Load Data

```python
import pandas as pd
full_data = pd.read_csv("../data/intermediate/forecasts_and_statuses.csv")
full_data.head()
```

Filter out actual status cols:

```python
selected_cols = [col for col in full_data.columns 
                 if not col.startswith(("Unnamed", "PWH_", "PWG_", "CWWH_", "CWWG_", "CWE_"))]
selected_cols
```

```python
selected_data = full_data[selected_cols]
selected_data = selected_data[selected_data["Date"] < "2024-01-01"]
selected_data.tail()
```

Number of occurrences by status type:

```python
for c in ["ACON", "MGEN", "OP41", "OP42", "OP43", "OP44", "OP45", "Abnormal"]: 
  print(selected_data[c].value_counts()) 
```

Majority are ACON (MCC/2).

### Delete NA Cols

```python
print(selected_data.isna().sum().sort_values(ascending=False))
for col in selected_data.columns:
   if col.startswith("AREG_"): 
     del selected_data[col]
print(selected_data.isna().sum().sort_values(ascending=False))
```

```python
selected_data.head(10)
```

### 1 Day Ahead

```python
selected_1_day_cols = [col for col in selected_data.columns
                       if not col.endswith(("_2", "_3", "_4", "_5", "_6"))]
selected_1_day = selected_data[selected_1_day_cols]
selected_1_day[selected_1_day.isna().any(axis=1)]
```

```python
selected_1_day = selected_1_day.dropna()
selected_1_day.to_csv("../data/processed/clean_1_day_ahead.csv", index=False)
```
